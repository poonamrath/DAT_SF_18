Since the weights don't vary much for tight ends, maybe we can just fill those values with the average weight for tight ends! 

We'll use the panadas [fillna() method](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.fillna.html)

```python
#store the value we want to use in a variable for later
te_avg_weight = df['weight'][df.position == 'TE'].mean()
```

```python
#Whats wrong with this statment? 
df.fillna(value=te_avg_weight, inplace=False) #inplace=True says actually perform the operation and save it
```

```python
#The above statement will replace ALL na values across the entire DF - which is not what we wanted
df['weight'].fillna(value=te_avg_weight, inplace=True)
```

Next, lets say we want to do an analysis on only players who were picked. To make the data cleaner, lets remove the players that have null values for the pick

We'll use the pandas [dropna() method](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.dropna.html)

```python
#Note the difference between axis = 1 and axis = 0
df.dropna(subset=['pick'], inplace=True, axis=1) # axis=1 will drop row with NA not columns for which axis =0
```

Sometimes we need to convert a categorical variable into a series of true/false variables. This is called a dummy variable. Let's use get_dummies() to create some dummy variables for the position column.

```python
pd.get_dummies(df.position).head(2)
```

Merge the dummy data with the original data set

```python
dummies = pd.get_dummies(df.position)
df = pd.concat([df,dummies])
df.head()
```
