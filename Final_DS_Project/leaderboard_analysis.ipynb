```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_board = pd.read_csv('./prudential-life-insurance-assessment_public_leaderboard.csv', index_col = 0)
```

```python
plt.figure(figsize=(9,6))

with plt.style.context('fivethirtyeight'):
    sns.distplot(df_board.Score)
    plt.title('Leaderboard Scores: Poonam/Highest = 0.65136/0.68107')
plt.show()
```
